# **Electric Vehicle Population Analysis Using Python**

# **Data Cleaning,Data Transformation and Feature Engineering**

```python
# Check duplicate rows
print("Duplicate Rows :", df.duplicated().sum())
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Fill Categorical Columns

cat_cols = df. select_dtypes(include='object').columns
for col in cat_cols:
   df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Fill Numerical Columns

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Drop Unwanted Columns
df.drop(columns=['Census_Tract'], inplace=True, errors='ignore')

df.head()
```

```python
df.drop(columns=['Vehicle_Location'], inplace=True, errors='ignore')

df.head()
```

```python
# Rename Column Names
df.rename(columns={
    "VIN (Vehicle Identification Number)(1-10)":"VIN",
    "Clean Alternative Fuel Vehicle (CAFV) Eligibility":"CAFV_Eligibility",
    "Electric Vehicle Type":"EV_Type",
    "Electric Range":"Electric_Range",
    "Postal Code":"Postal_Code",
    "Model Year":"Model_Year",
    "Legislative District":"Legislative_District",
    "DOL Vehicle ID":"Vehicle_ID",
    "Electric Utility":"Electric_Utility",
   }, inplace=True)

df.columns
```

```python
# Replace Column Names
df["CAFV_Eligibility"] = df["CAFV_Eligibility"].replace({
    "Clean Alternative Fuel Vehicle Eligible": "Eligible",
    "Not eligible due to low battery range": "Not Eligible",
    "Eligibility unknown as battery range has not been researched": "Unknown"
})
```

```python
print(df["CAFV_Eligibility"].value_counts())
```

```python
# Convert Data Types
df['Postal_Code'] = df['Postal_Code'].astype(int)
df['Legislative_District'] = df['Legislative_District'].astype(int)

df.info()
```

# **Feature Engineering**

```python
# Vehicle Age
current_year = 2026

df['Vehicle_Age'] = current_year - df['Model_Year']
```

```python
# Range Category
df['Range_Category'] = pd.cut(
    df['Electric_Range'],
    bins=[0,100,250,1000],
    labels=['Low','Medium','High']
)
```

```python
# EV Category
df['EV_Category'] = np.where(
    df['Electric_Range']>=250,
    'Long Range',
    'Standard Range'
)
```

```python
df.isnull().sum()
```

```python
df[df["Range_Category"].isnull()]
```

```python
# impute null values
def range_category(x):
    if x == 0:
        return "Unknown"
    elif x <= 100:
        return "Short Range"
    elif x <= 250:
        return "Medium Range"
    else:
        return "Long Range"

df["Range_Category"] = df["Electric_Range"].apply(range_category)
```

```python
df.isnull().sum()
```

# **Statistical analysis & Visualizations**

```python
# MEASURE OF CENTRAL TENDENCY + Chart
num_cols = df.select_dtypes(include=['int64','float64']).columns

central_tendency = pd.DataFrame({
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Mode': df[num_cols].mode().iloc[0]
})

print(central_tendency)
```

```python
# visualization
for col in num_cols:

    plt.figure(figsize=(4,5))

    sns.histplot(df[col], bins=30, kde=True)

    plt.axvline(df[col].mean(), color='red', linestyle='--',
                label=f"Mean = {df[col].mean():.2f}")

    plt.axvline(df[col].median(), color='green', linestyle='-',
                label=f"Median = {df[col].median():.2f}")

    plt.title(f"Distribution of {col}")

    plt.legend()

    plt.show()
```

**1. Histogram (Mean & Median)**

**Interpretation (Symmetric Distribution)**

The mean and median are very close, indicating the data is approximately normally distributed.
Most observations are concentrated around the center.
There is little evidence of extreme values affecting the distribution.

**Interpretation (Right-Skewed Distribution)**

The mean is greater than the median.
A few high values pull the mean toward the right.
Most observations are concentrated in the lower range.

**Interpretation (Left-Skewed Distribution)**

The mean is less than the median.
Some unusually low values pull the mean toward the left.
Most observations are concentrated in the higher range.

```python
# Variance and Standard Deviation
variation = pd.DataFrame({
    'Variance': df[num_cols].var(),
    'Standard Deviation': df[num_cols].std()
})

print(variation)
```

```python
# visualization
variation.plot(kind='bar',
               figsize=(10,5))

plt.title("Variance and Standard Deviation")

plt.ylabel("Values")

plt.xticks(rotation=45)

plt.show()
```

**2. Variance and Standard Deviation**

**Interpretation**

**Census Tract** has the **highest variance and standard deviation**, indicating a very large spread in its values. This is expected because it is an identification/geographical code with large numeric values rather than a continuous measurement.

**DOL Vehicle ID** also shows relatively **high variance** and **standard deviation** compared to the other variables. Since it is a unique vehicle identifier, its values naturally vary widely and do not represent a meaningful business metric.

**Postal Code, Model Year, Electric Range,** and **Legislative District** have much **lower variance** and **standard deviation** than Census Tract. Their values are comparatively less dispersed around the mean.

The **extremely large variance** of **Census Tract** dominates the chart, making the variation in the other variables appear very small. This occurs because variance is highly influenced by the scale of the data.

```python
# Skewness and Kurtosis
shape = pd.DataFrame({
    'Skewness': df[num_cols].skew(),
    'Kurtosis': df[num_cols].kurt()
})

print(shape)
```

```python
# visualization
shape.plot(kind='bar',
           figsize=(10,5))

plt.title("Skewness and Kurtosis")

plt.xticks(rotation=45)

plt.show()
```

**3. Interpretation of the Skewness and Kurtosis Chart**

The chart compares the Skewness and Kurtosis values of the numerical variables in the Electric Vehicle (EV) dataset.

**Postal Code** shows a **high positive kurtosis** and noticeable skewness, indicating that the values are highly concentrated with a few extreme observations. This variable is not normally distributed and represents geographical identifiers rather than continuous numerical data.

**Census Tract** has the **highest kurtosis** among all variables and a significant negative skewness, suggesting the presence of extreme values and a highly peaked distribution. Since it is an identification field, these statistics are not meaningful for business analysis.

**Model Year, Electric Range, Legislative District,** and **DOL Vehicle ID** have skewness and kurtosis values close to zero compared to the other variables, indicating relatively less asymmetry and fewer extreme values.

The extremely high kurtosis values for **Postal Code** and **Census Tract** imply that these columns contain many repeated values with a few rare observations. As these are categorical/identifier fields, they should not be interpreted as continuous numerical variables in statistical analysis.

```python
# box plot outlier detection
for col in num_cols:

    plt.figure(figsize=(7,4))

    sns.boxplot(x=df[col])

    plt.title(f"Box Plot of {col}")

    plt.show()
```

**Interpretation of Box Plots**

The box plots were used to identify the presence of outliers in the numerical features of the Electric Vehicle dataset. Outliers are represented by individual points lying outside the whiskers of the box plot.

Electric Range contains several outliers, indicating that a few vehicles have unusually high or low electric ranges compared to the majority of vehicles.
Model Year shows a relatively compact distribution with very few or no significant outliers, indicating that most vehicles belong to recent model years.

Postal Code, Legislative District, DOL Vehicle ID, and Census Tract also display apparent outliers. However, these columns are identifier or location-based variables, so the detected outliers are expected and do not necessarily represent data errors.

Overall, the box plots reveal that only continuous numerical variables should be considered for meaningful outlier analysis, while identifier fields should generally be excluded.
